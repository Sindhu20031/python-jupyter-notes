# (7) NumPy Vectorization


#### - NumPy vectorization means performing operations on an entire array at once without using loops.

#### Why Vectorization is Used

Faster than loops

Less code

More readable

Optimized in NumPy

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
result = arr + 5

print(result)
```

### Explanation

5 is added to each element

No loop is written

Operation happens internally in C (fast)

### Vectorization vs Loops
❌ Using Loop

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
result = []

for i in arr:
    result.append(i + 5)

print(result)
```

### ✅ Using Vectorization

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
print(arr + 5)
```

# Parallel Computing in NumPy
##### Parallel computing means doing multiple calculations at the same time instead of one by one.

#### NumPy internally uses:

C language

Multi-core CPU

SIMD (Single Instruction Multiple Data)

So even one NumPy line runs faster using parallelism.

```python
import numpy as np

a = np.random.rand(1000000)
b = np.random.rand(1000000)

c = a + b   # runs in parallel internally
print(c)
```

# View
####  A view is just another way to see the same array, so changing it also changes the original array.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = a.view()

b[0] = 100

print(a)  # original array
print(b)  # view array
```

# Copy
##### A Copy creates a completely new array with its own memory.
##### So, changing the copy will NOT affect the original array.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
c = a.copy()

c[0] = 500

print(a)  # original array
print(c)  # copied array
```

# (8) Descriptive Statistics Using NumPy

#### Descriptive statistics means summarizing data using numbers so we can easily understand it.

## 1) Count
- Count tells how many values are present in the data.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

count = len(data)

print("Data:", data)
print("Count:", count)
```

## 2) Sum
- Sum means the total of all values.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

total = np.sum(data)

print("Data:", data)
print("Sum:", total)
```

## 3) Mean (Average)

- Mean is the average value of the data.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)

print("Data:", data)
print("Mean:", mean)
```

## 4) Median

- Median is the middle value when data is sorted.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

median = np.median(data)

print("Data:", data)
print("Median:", median)
```

## 5) Variance
- Variance shows how spread out the data is.

## 6) Standard Deviation

- Standard deviation shows how far values are from the mean.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

std_dev = np.std(data)

print("Data:", data)
print("Standard Deviation:", std_dev)
```

## 7) Percentile

- Percentile shows the value below which a percentage of data lies.
- A percentile tells you the value below which a certain percent of the data falls.

#### Example in words:
- If the 50th percentile is 30, it means 50% of the values are less than or equal to 30.
- If I say 25th percentile, it means one-quarter of the numbers are below that value.
#### Percentile = position of a value in the data.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

p50 = np.percentile(data, 50)

print("Data:", data)
print("50th Percentile:", p50)
```

- Covariance shows whether two variables move together.

# (5) Groupby, Pivot table and Cross Tab

## 1. GROUPBY
### What is GroupBy?
- GroupBy is used to group data based on a column and apply calculations like sum, mean, count.

### 1.1 GroupBy with Mean

```python
df.groupby("Loan_Status")["ApplicantIncome"].mean()  # groups data by Loan_Status and finds average income
```

# 2. PIVOT TABLE
### What is Pivot Table?
- A pivot table is a method used to organize, group, and summarize large amounts of data into a simple table, so we can easily compare values and understand the data.

## 2.1 Simple Pivot Table

```python
pd.pivot_table(
    df,                                # dataset
    values="LoanAmount",               # values to calculate
    index="Loan_Status",               # rows
    aggfunc="mean"                     # calculation
)
```

### 2.2 Pivot Table with Rows & Columns
- It groups the data by Gender (rows) and Loan_Status (columns) and shows the average LoanAmount in a table.
- Show average loan amount for each gender, separately for approved and not approved loans.

```python
pd.pivot_table(
    df,                                # dataset
    values="LoanAmount",               # value column
    index="Gender",                    # rows
    columns="Loan_Status",             # columns
    aggfunc="mean"                     # aggregation
)
```

### 2.3 Pivot Table with Multiple Functions
- It is used to summarize the same column using more than one calculation (like mean, max, count) at the same time, grouped by a category.

```python
pd.pivot_table(
    df,
    values="ApplicantIncome",
    index="Education",
    aggfunc=["mean", "max", "count"]
)
```

# 3. CROSS TAB (CROSSTAB)
### What is Crosstab
- A crosstab shows how many times each combination of two categories appears in the data.

### 3.1 Simple Crosstab
- Count how many males and females got loan approved and not approved

```python
pd.crosstab(df["Gender"], df["Loan_Status"])  
```

### 3.2 Crosstab with Normalization
### i) normalize=True 
- Show loan approval percentages for each property area.
#### normalize=True → percentages, not counts

### Memory Trick 

normalize=True → overall percentages

normalize="index" → row-wise percentages

normalize="columns" → column-wise percentages

```python
pd.crosstab(df["Property_Area"], df["Loan_Status"], normalize=True)  # shows percentage instead of count
```

```python
pd.crosstab(df["Gender"], df["Loan_Status"], normalize=True)
# percentage out of total records
```

### ii). normalize="index" (ROW-WISE)
- It shows percentages inside each row, meaning for each gender, how many got loan approved and not approved.

```python
pd.crosstab(df["Gender"], df["Loan_Status"], normalize="index")
# shows loan approval % for each gender
```

### iii). normalize="columns" (COLUMN-WISE)
- It shows percentages inside each column, meaning for approved and not approved loans, how many are male and female.

```python
pd.crosstab(df["Gender"], df["Loan_Status"], normalize="columns")  # shows gender distribution within each loan status
```

# (6) Concatenating and Merging Joining

### 1.1 Concatenation (concat)

- Concatenation means combining DataFrames either row-wise or column-wise.

### i) Row-wise Concatenation
- Row-wise concatenation is used to combine data from multiple tables with the same columns into one table

```python
df_row = pd.concat([df, df])
print(df_row)
```

```python
df_row.head()
```

### ii) Column-wise Concatenation
- Put ApplicantIncome and LoanAmount columns next to each other in one table.
- axis=1 means add as columns, not rows

```python
df_col = pd.concat([df[["ApplicantIncome"]], df[["LoanAmount"]]], axis=1)
print(df_col)
```

### 1.2 Merging / Joining (merge)
- Merging means connecting two tables using one common column so related data comes together in one table.

### Create second DataFrame (example)
- Create a new table with only Loan_ID and Property_Area columns from the original data.

```python
df_extra = df[["Loan_ID", "Property_Area"]]
print(df_extra)
```

## 1) Inner Join
- Keeps only the rows that match in both tables.
#### The "how" parameter is used to decide which rows should be kept when two tables are merged.

```python
pd.merge(df, df_extra, on="Loan_ID", how="inner")
```

### 2) Left Join
Keeps all rows from the left table and only matching rows from the right table.

```python
pd.merge(df, df_extra, on="Loan_ID", how="left")
```

### 3) Right Join
- Keeps all rows from the right table and only matching rows from the left table.

```python
pd.merge(df, df_extra, on="Loan_ID", how="right")
```

### 4) Outer Join
- Keeps all rows from both tables, whether they match or not.

```python
pd.merge(df, df_extra, on="Loan_ID", how="outer")
```

### 1) Convert Column to DateTime (to_datetime)
Converts a column into DateTime format.

```python
import pandas as pd

df = pd.DataFrame({"Loan_Date": ["2024-01-10"]})
df["Loan_Date"] = pd.to_datetime(df["Loan_Date"])
print(df)
```

### 2) Extract Year


Extracts year from date.

```python
import pandas as pd

df = pd.DataFrame({"Loan_Date": ["2024-01-10"]})
df["Loan_Date"] = pd.to_datetime(df["Loan_Date"])
print(df["Loan_Date"].dt.year)
```
