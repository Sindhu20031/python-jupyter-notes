# Exceptional handling
- An exception is an error that happens while your program is running.
- It stops the program if you don’t handle it.
- Exception handling means catching and fixing those errors using special code.
- So that your program doesn’t crash and shows a friendly message instead.

## Syntax:-

try:
    # code that might raise an exception
except ExceptionType:
    # code to handle the exception
else:
    # code that runs if no exception occurs
finally:
    # code that always runs, regardless of exceptions

```python
a = 10
b = 0
print(a / b)   # This line will cause an error
```

```python
a = 10
b = 5
c= 7
try:
    print(c)
except:
    print("error")
d= a+b
print(d)
```

```python
a = 10
b = 5
c= 7
try:
    print(c)
except:
    print("error")
else:
    print("No error")
d= a+b
print(d)
```

```python
a = 10
b = 5
c= 7
try:
    print(c)
except:
    print("error")
else:
    print("No error")
finally:
    print("Done")
d= a+b
print(d)
```

# Types Of Errors

## 1) ZeroDivisionerror:-

## eg1:-

### Division with error check

```python
a = 4
b = 0
try:
    print(a/b)
except:
    print("cannot divide by zero!")
print("Done with division check")
```

```python
a = 4
b = 0
try:
    print(a/b)
except:
    print("Proceed")
a+b
```

```python
a = 10
b = 0

try:
    result = a / b  #  This will cause ZeroDivisionError
    print("Result is:", result)
except ZeroDivisionError:
    print("You cannot divide by zero!")
else:
    print("Division successful!")
finally:
    print("This block always runs (whether error or not)")
```

# 2) ValueError:-
- It happens when you give a wrong value to a function — especially when converting data types.
- eg:- You gave the right data type (like string), but the value inside it is not okay.

```python
num = int("abc")  #  'abc' is not a valid integer
print(num)
```

```python
try:
    num = int("abc")  # This will cause ValueError
    print(num)
except ValueError:
    print("Please enter a valid number!")
```

```python
try:
    num = int("abc")  # This will cause ValueError
    print(num)
except ValueError:
    print("Please enter a valid number!")
else: 
    print(f"Converted into value")
finally:
    print("Done")
```

# 3) TypeError:-
- Raised when an operation is aplied to an object of inappropriate type.
- TypeError happens when you try to do an operation between the wrong data types (like adding string and number).

```python
"2" + 2
```

```python
a = "10"
b = 5
print(a + b)           #  Can't add string and number
```

```python
a = "10"
b = 5
try:
    print(a + b)
except TypeError:
    print("You cannot add a string and a number!")
```

# 4) IndexError:-
- You get an IndexError when you try to access a position (index) in a list that doesn’t exist.

```python
lst = [1, 2, 3]
print(lst[5])
```

```python
fruits = ["apple", "banana", "mango"]

try:
    print(fruits[5])
except IndexError:
    print("That index does not exist in the list!")
```

```python
fruits = ["apple", "banana", "mango"]

try:
    print(fruits[1])                    #  This is a valid index
except IndexError:
    print("Invalid index!")
else:
    print("Item printed successfully.")
finally:
    print("This always runs.")
```

## 5) KeyError:-
- You get a KeyError when you try to access a key in a dictionary that doesn’t exist.

```python
student = {"name": "Sindhu", "age": 22}
print(student["branch"])                        #  'branch' key doesn't exist
```

##### Catching with try-except:

```python
student = {"name": "Sindhu", "age": 22}

try:
    print(student["branch"])
except KeyError:
    print("That key does not exist in the dictionary!")
```

##### Using try, except, else, finally:

```python
student = {"name": "Sindhu", "age": 22}

try:
    print(student["age"])  
except KeyError:
    print("Key not found!")
else:
    print("Key accessed successfully.")
finally:
    print("This always runs.")
```
