```python
a = 6
print(~a)     # -(a+1)
```

# right shift (>>)


```python
if we want to do 1 shift it is x>>1 ---------------->>>>> 1 represnts shifts(like how many shifts)
- x>>2 ---------------------->>> 2 shifts
- x>>3 ----------------------->>> 3 shifts
- 0  0  1  1
-    0  0  1  ------>> 1 shifts
-       0  0 ------->> 2 shifts
```

```python
x = 3
print(x>>1)   
print(x>>2)
print(x>>3)
```
