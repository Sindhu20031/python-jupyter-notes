# Power BI

- Power BI is a business analytics tool developed by Microsoft that allows you to visualize data, share insights, and make informed business decisions. In simple terms, it helps you turn raw data into interactive reports and dashboards that are easy to understand.
- Here’s a breakdown:

Data Connection: Power BI can connect to many data sources—Excel, SQL databases, cloud services, APIs, etc.

Data Transformation: Using Power Query, you can clean, filter, and transform your data to make it ready for analysis.

Data Modeling: You can create relationships between tables and define calculations using DAX (Data Analysis Expressions).

Visualization: You can create charts, graphs, maps, and interactive dashboards that update dynamically with your data.

Sharing & Collaboration: Reports and dashboards can be published to the Power BI service for others to view and interact with online.

Use cases:

Tracking sales performance

Monitoring KPIs (Key Performance Indicators)

Financial reporting and forecasting

Analyzing customer behavior

Think of it as a tool that turns complex data into visual stories that anyone in a company can understand.

### Why Power BI

- Pandas, Matplotlib, and Plotly are coding tools. You need to write Python code to clean data, make charts, and analyze it. They are great for programmers, but if someone doesn’t know coding, it’s hard to use.

Power BI, on the other hand:

- Is drag-and-drop: You can make charts and dashboards by clicking, not coding.

- Is interactive: You can click filters, drill down, and explore data live.

- Can connect to many data sources easily.

- Lets you share dashboards with your team or management online.

# What is Dashboard?
- A dashboard in Power BI is a one-page summary that shows important data using charts and visuals — all in one place.

#### Multiple visualizations in Power BI means showing different types of charts and visuals together — like bar charts, pie charts, line graphs, and maps — on one report or dashboard.
- Purpose: To see data from many angles and understand patterns better.

#### Example:
For sales data, you might use —

A bar chart for sales by product,

A map for sales by region,

A line chart for monthly trends.

- All these together = multiple visualizations in one report.

## Components of Power BI
1) Power BI Desktop – The main app where you create reports and visuals on your computer.
2) Power BI Service – The online (cloud) platform where you share and view dashboards.
3) Power BI Data Gateway – Connects your local data (like Excel or SQL) to Power BI Service.
4) Power BI Mobile App – Lets you view reports and dashboards on your phone/tablet.
5) Power BI Report Builder – Used to create paginated (print-style) reports.
6) Power BI Report Server – For companies that want to host reports on their own servers, not in the cloud.

#### In short:

Desktop → Create

Service → Share

Gateway → Connect data

Mobile → View anywhere

Report Builder/Server → For advanced or offline use

## Power BI Workflow or Power BI Working Process
- It means the step-by-step process of how Power BI works — from getting data to sharing insights.
1) Data → Bringing in data from sources (like Excel, SQL, etc.).
2) Data Transformation → Cleaning and shaping the data (removing errors, changing columns, etc.).
3) Data Modeling → Connecting tables and creating relationships (like linking “Sales” and “Customers”).
4) Reports & Dashboards → Creating charts, visuals, and summaries to show insights.
5) Publishing → Uploading your reports to the Power BI Service (online) to share with others.

##### In short:
Get data → Clean it → Connect it → Visualize it → Share it.

## What is Visualization in Power BI?

Visualization = The way we show data graphically so it’s easy to understand.

Examples: charts, graphs, maps, tables, KPIs.

Goal: Make numbers meaningful & easy to read.

#### Types of Visualizations in Power BI
A. Basic Charts

#### Bar Chart

Shows comparison between categories.

Horizontal bars.

#### Column Chart

Similar to bar chart but vertical bars.

Good for comparing quantities over time.

#### Line Chart

Shows trend over time.

Connects data points with a line.

#### Area Chart

Like line chart but fills area below the line.

Good for showing total over time.

#### Pie Chart

Shows percentage of a whole.

Each slice = part of total.

#### Donut Chart

Like pie chart but center is empty.

Looks cleaner with labels.

#### 1. Card

What it is: A Card shows only one single important number.

Use: When you want to highlight one key value like total sales, total customers, or total profit.

###### Example:

Total Sales = 50,000

Total Customers = 1,200

Visual in Power BI: Big number shown on the screen, easy to see.

#### 2. Multi-row Card

What it is: A Multi-row Card shows many numbers together.

Use: When you want to show multiple important metrics in one place.

###### Example:

Total Sales: 50,000
Total Customers: 1,200
Total Orders: 3,500


Visual in Power BI: Lists multiple numbers one below the other.

#### 3. KPI (Key Performance Indicator)

What it is: KPI shows how close you are to your target.

Use: To track progress, like if you are achieving your goal or not.

###### Example:

Target Sales = 60,000

Current Sales = 50,000 → 83% of target achieved

Indicator shows up arrow if good, down arrow if bad.

Visual in Power BI: Shows current value, target, and trend in one visual.

### 1. Table Visual

What it is:

A table is like a normal Excel table.

It shows data row by row.

Best when you want to see detailed information.

##### Example:

Columns: Product Name, Price, Quantity

Rows: Each product in your store

You can sort by Price or Quantity.

Use Case:

See all products with their prices and sales.

### 2. Matrix Visual

What it is:

Like a table but can show grouped data.

You can put rows and columns and see totals.

Can drill down to see more details.

##### Example:

Rows: Region

Columns: Month

Values: Sales

You can see sales by Region and Month, and totals for each region or month.

Use Case:

Compare sales of regions month by month.

### 3. Scatter Chart

What it is:

Shows points on a graph based on 2 (or 3) numbers.

X-axis: one number, Y-axis: another number

Optional size or color of dots for extra info.
- Scatter plot shows relationship between two numerical variables

##### Example:

X-axis: Advertising Budget

Y-axis: Sales

Each dot = a product

Bigger dot = more profit

Use Case:

See relationship between two things, like budget vs sales.

- Dots make a trend line
- You can see positive relationship easily

### Waterfall Chart

- Shows how a starting value changes step by step to reach a final value.
- Each step = increase or decrease.
- Good for profit/loss, revenue changes, or cost analysis.

#### Example

You have a company starting with $1000, then:

Revenue + $300

Cost - $150

Bonus + $50

Waterfall shows how $1000 → $1200 step by step

- Bars go up or down depending on + or -
- Final bar = total amount

### 1. Map Chart

Shows data on a map.

Each location = dot or color.

Bigger dot / darker color = bigger value.

Use it to see sales, population, or any numbers by location.

Example:

Sales in different cities → big dot in Mumbai, small dot in Pune.

### 2. Tree Map (Box inside Box)

Shows data as boxes inside boxes.

Bigger box = bigger value.

Good to see part-to-whole relationships.

Example:

Category = Electronics (big box)

Products inside = TV, Phone (smaller boxes)

Shows which product sells more.

### Slicer
- A slicer is like a filter button for your report.
- It helps you see only the data you want.
- Click something in slicer → all charts update automatically.

### Example
- You have sales data for all months: Jan, Feb, Mar…
- Add a slicer for Month
- Click Feb → charts only show February data
##### Slicer = interactive filter
##### Makes your dashboard dynamic

### Ribbon Chart 

Shows how categories change rank over time.

Each “ribbon” = a category.

Width shows value, position shows ranking.

Good to see who is leading and who is falling over months or years.
