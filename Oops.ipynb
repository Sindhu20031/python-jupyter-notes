### Start with empty basket add some 3 items and whenever they are doing scan for every product scan it should display price of each item it should add with next item print final bill

```python
class ShoppingCart:
    def __init__(self):
        self.basket={}
        self.total=0
    def scan_item(self, item, price):
        self.basket[item] = price
        self.total+= price
        print(f"Scanned: {item} - rps{price}")
    def print_bill(self):
        print("\n Final Bill:")
        for item, price in self.basket.items():
            print(f"{item}: rps{price}")
        print(f"\nTotal: rps{self.total}")
cart = ShoppingCart()

# Add 3 items to the basket
cart.scan_item("Milk", 30)
cart.scan_item("Bread", 25)
cart.scan_item("Eggs", 50)

# Print final bill
cart.print_bill()      
```

### Binary to Decimal using loop

```python
binary = "100111"
decimal = 0
power = 0

for digit in reversed(binary):    
    if digit=='1':
        decimal+=2**power
    power +=1
print("Decimal:", decimal)
```

## 2) Multilevel Inheritance:-
- Multilevel inheritance means one class inherits from another class, and then another class inherits from that - --- second class — like a chain.
- Grandfather → Father → Child

## eg1:-

```python
class Grandfather:
    def home(self):
        print("This is Grandfather's home")

class Father(Grandfather):
    def car(self):
        print("This is Father's car")

class Child(Father):
    def bike(self):
        print("This is Child's bike")

c = Child()
c.home()   # From Grandfather
c.car()    # From Father
c.bike()   # From Child
```

```python
class parent:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        final = self.x + self.y
        return final

    def mul(self):
        final = self.x * self.y
        return final

class child(parent):
    def __init__(self, x, y):
        parent.__init__(self, x, y)

    def subt(self):
        sub = self.x - self.y
        return sub

class child1(child):
    def __init__(self, x, y):
        child.__init__(self, x, y)

p2 = child1(11, 10)
print(p2.add())   
print(p2.subt())  
print(p2.mul())     
```

## 3) Multiple Inheritance:-
- Multiple inheritance means one class can take code (like functions or variables) from two or more classes.
It’s like learning from both parents instead of just one.
- If a child learns from both mother and father, that’s multiple inheritance.

```python
class Mother:
    def cook(self):
        print("Mother can cook")

class Father:
    def drive(self):
        print("Father can drive")

class Child(Mother, Father):  # Inheriting from both
    def play(self):
        print("Child can play")

c = Child()
c.cook()   # From Mother
c.drive()  # From Father
c.play()   # Own method
```

```python
class parent:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        final = self.x + self.y
        return final

class child:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def subt(self):
        sub = self.x - self.y
        return sub

class Grandchild(parent, child):   # Multiple inheritance
    def __init__(self, x, y):
        parent.__init__(self, x, y)  # we can call one init (same x, y used in both)

    def mul(self):
        final = self.x * self.y
        return final

p2 = Grandchild(11, 10)
print(p2.add())    # from parent - 21
print(p2.subt())   # from child  -1
print(p2.mul())    # from Grandchild -110
```

```python
class Father:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        final = self.x + self.y
        return final

class Mother:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def subt(self):
        sub = self.x - self.y
        return sub
    
class Child(Father, Mother):                 # child class using mutiple inheritance
    def __init__(self, x, y):              ### calling both parent constructors manually
        Father.__init__(self, x, y)
        Mother.__init__(self, x, y)
    
    def mul(self):
        final = self.x* self.y
        return final
         

c = Child(11, 10)
print(c.add())    # From Father 
print(c.subt())   # From Mother 
print(c.mul())
```

## 4) Hierarchical inheritance:-
- Hierarchical inheritance means one parent class is shared by multiple child classes.
- It’s like one teacher teaching many students—all students learn from the same teacher (parent class).


```python
# Parent class
class Animal:
    def sound(self):
        print("Animals make sounds")

# Child class 1
class Dog(Animal):
    def speak(self):
        print("Dog says: Woof")

# Child class 2
class Cat(Animal):
    def speak(self):
        print("Cat says: Meow")

# Create objects of child classes
d = Dog()
c = Cat()

# Call parent and child methods
d.sound()   # from Animal
d.speak()   # from Dog

c.sound()   # from Animal
c.speak()   # from Cat
```

```python
class Father:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        final = self.x + self.y
        return final

class Son1(Father):   # Inherits from Father
    def subt(self):
        sub = self.x - self.y
        return sub

class Son2(Father):   # Inherits from Father
    def mul(self):
        final = self.x * self.y
        return final

# Create object for Son1 and call add and subt
s1 = Son1(11, 10)
print(s1.add())   # From Father
print(s1.subt())  # From Son1   

# Create object for Son2 and call add and mul
s2 = Son2(11, 10)
print(s2.add())   # From Father 
print(s2.mul())   # From Son2   
```

```python
class Addition:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        final = self.x + self.y
        return final

class Subtraction(Addition):   # Inherits from Addition
    def subt(self):
        sub = self.x - self.y
        return sub

class Multiplication(Addition):   # Inherits from Addition
    def mul(self):
        final = self.x * self.y
        return final

# Object for Subtraction class
s1 = Subtraction(11, 10)
print(s1.add())                                
print(s1.subt())                              

# Object for Multiplication class
s2 = Multiplication(11, 10)
print(s2.mul())               # From Multiplication 
```

# 2) Encapsulation:-
- Encapsulation means hiding the data inside a class and only allowing access to it through methods (functions).
- It protects the data from being changed directly.

```python
class Student:                        # Creating a class named Student
    def __init__(self):              # Constructor method that runs when object is created
        self.__marks = 0            # Private variable (cannot be accessed directly outside the class)

    def set_marks(self, m):         # Method to set marks (takes one argument m)
        if 0 <= m <= 100:           # Check if marks are in valid range
            self.__marks = m        # If yes, set the private variable
        else:
            print("Invalid marks")  # If not, show an error message

    def get_marks(self):            # Method to return the marks
        return self.__marks         # Gives back the value of private marks

# Outside the class

s = Student()                       # Create object 's' of class Student
s.set_marks(85)                     # Call set_marks method to store 85 as marks
print(s.get_marks())               # Call get_marks to print the stored marks → 85

s.set_marks(150)                    # Trying to set invalid marks
print(s.get_marks())               # Still prints 85, because 150 was rejected
```

```python
class BankAccount:                      # Creating a class named BankAccount
    def __init__(self):                 # Constructor method
        self.__balance = 0             # Private variable to store account balance

    def deposit(self, amount):         # Method to add money
        if amount > 0:                 # Check if amount is positive
            self.__balance += amount  # Add amount to balance
        else:
            print("Invalid deposit")   # If negative amount, show error

    def withdraw(self, amount):        # Method to withdraw money
        if 0 < amount <= self.__balance:  # Check if enough balance
            self.__balance -= amount     # Subtract amount from balance
        else:
            print("Insufficient funds or invalid amount")  # Show error

    def get_balance(self):             # Method to check balance
        return self.__balance          # Return the balance value

# Outside the class
acc = BankAccount()                    # Create a BankAccount object named acc
acc.deposit(500)                       # Deposit ₹500
acc.withdraw(200)                      # Withdraw ₹200
print(acc.get_balance())              # Check remaining balance → ₹300

acc.withdraw(400)                      # Try to withdraw ₹400 (more than balance)
print(acc.get_balance())              # Balance is still ₹300, as withdrawal failed
```

## 3) Abstraction:-
- Abstraction in Python means hiding complex things and only showing what is needed.
- It helps you focus on what an object does instead of how it does it.

```python
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass

# Child class
class Car(Vehicle):
    
    def start(self):
        print("Car engine started.")

# Create object of child class
c = Car()
c.start()
```

```python
from abc import ABC, abstractmethod

class Phone(ABC):
    
    @abstractmethod
    def call(self):
        pass

class Android(Phone):
    
    def call(self):
        print("Calling from Android phone...")

class iPhone(Phone):
    
    def call(self):
        print("Calling from iPhone...")

# Use
a = Android()         #### Phone is abstract → you cannot create its object.
a.call()

b = iPhone()
b.call()
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    
    def sound(self):
        print("Dog says: Woof!")

class Cat(Animal):
    
    def sound(self):
        print("Cat says: Meow!")

# Use
d = Dog()
d.sound()

c = Cat()
c.sound()
```
