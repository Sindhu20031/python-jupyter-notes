# Advanced Data Type-List:-

 - List is ordered collection of items
 - represented by square brackets----[]
 - List allow all data types(int, float, bool, mixed data types, nested list)
 - List is mutable data type (we can add, delete, update)
 - It also allows duplicates (means we can write same thing n number of times)

### Eg:- Create empty list

```python
l1=[]
print(l1)
print(type(l1))
```

### Integer list

```python
l2=[1, 2, 4, 5, 6, 8]
print(l2)
```

### String List

```python
l3=['sindhu', 'good', 'girl']
print(l3)
```

### Float List

```python
l4=[3.4, 5.9]
print(l4)
```

### Mixed List

```python
l5=[4.5, 'data', 9, True]
print(l5)
```

### nested list

```python
l6=[5, 7, ['sindhu', True], 7.8, ['data', 6]]
print(l6)
print(l6[3:9])
print(l6[1:-6])
print(l6[:: -9])
print(l6[3::9])
```

```python
print(len(l6))
```

# List Methods/built-in Function:-

### 1) append:-
#### is used to add single element to the list

```python
ls=[1, 4, 6]
ls.append("sindhu")
print(ls)
```

### 2) extend:-
#### is used to extend existing list with more than one element

```python
ls=[3, 4, 6, "sindhu"]
ls.extend(["good", 6])
print(ls)
```

### 3) Changing values in a List:-

```python
ls=[3, 4, 5, 'sindhu', 8]
ls[3]=100
ls
```

### 4) insert:-
#### We can insert any values of index
#### Adds item at a specific position

```python
ls=[3, 4, 7, 'sindhu', 'good']
ls.insert(3, 'cat')
ls
```

### 5) Pop:-
#### remove elements from the list
#### Removes item by index (default last one)

```python
fruits=['apple', 'watermelon', 'orange']
fruits.pop(1)     ## removing with index
fruits
```

```python
fruits.pop()
```

### 6) remove:-
#### Removes the first occurrence of an item

```python
ls=[3, 4, 5,]
ls.remove(4)
ls
```

```python
ls=['sindhu', 'nelli', 'sindhu', 'param']
ls.remove('sindhu')
ls
```

### 7) delete:-
#### The del keyword in Python is used to delete elements from a list by index or slice.

```python
num=[1, 4, 6, 8, 9, 0, 5, 4, 3]
del num[1:5]            ## using slice
num
```

```python
names=['sindhu', 'nelli', 'param']
del names[2]            ## using index number
names
```

```python
num = [1, 2, 3, [10, 11, 12, 13, 14, 15]]
del num[3][4]                   # [3][4] means: access index 4 of the item at index 3 in a list.
print(num)                                        
```

### 8) joining two list:-
#### Joining two lists means combining their elements into one list using the + operator or .extend() method

```python
l1=["sindhu"]
l2=["good"]
l1+l2
```

### 11) reverse():-
#### reverses the order of elements in a list (changes the list itself).

```python
numbers = [1, 2, 3, 4]
numbers.reverse()
print(numbers) 
```

### 12) clear():-
#### clear will clears all the elements from the list then list will be empty.

```python
a = [1, 2, 3]
a.clear()
print(a)
```

### 13) copy()-
#### Returns a shallow copy of the list.
#### It makes a new list, but the inside parts are still connected to the old list.

##### eg:-
#### ---->>> original a=[2, 3, 4]
#### --->>> duplicate b = a.copy()

```python
a = [1, 2, 3]
b = a.copy()
print(b)
```

#### nested list 

```python
x=[1, 2, 3, [4, 5]]
y=x.copy()
y[3][0]=100
print(y)
```

```python
a = [1, 2, 3]
b = a.copy()
b[0] = 100
print(a)  
print(b)  
```

```python
####  in this code, we don’t need import copy because you're using copy(), which is a built-in list method.

a = [1, 2, 3, 4]          # original list
import copy               # import the copy module (needed only for deepcopy, not for copy())
b = a.copy()              # make a shallow copy of list 'a'
print(a)                  # prints original list: [1, 2, 3, 4]
print(b)                  # prints copied list:   [1, 2, 3, 4]
```

```python
b[2]=100
print(a)
print(b)
```

#### we are getting same for both so to avoid this condition we use deepcopy.

```python
a = [[1, 2], [3, 4]]
b = a.copy()
b[0][0] = 999      ### Syntax:- list_name[outer_index][inner_index] = new_value
print(a)  
print(b)      #### Changing b also changes a. Why? Because it didn't really copy the inner list  just linked to it.
```

### 14)  deepcopy():-
#### It copies everything, including all the lists inside lists. So changes in the new copy won’t affect the original.
#### It makes a completely new list, including all inside parts — nothing is connected to the old list.



```python
b = copy.deepcopy([[1, 2], [3, 4]]) 
print(b)
```

```python
x=[1, 2, 3, [4, 5]]
y=copy.deepcopy(x)
print(y)
```

```python
x=[1, 2, 3, [4, 5]]
y=copy.deepcopy(x)
x[3][0]=100
print(x)
print(y)
```

### 15) sort
#### sort() arranges the list items in order  by default, from small to big (ascending).

```python
a = [5, 2, 9, 1]
a.sort()
print(a)
```

```python
a.sort(reverse=True)    # decsending order
print(a)
```
