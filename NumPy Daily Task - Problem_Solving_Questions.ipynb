## 18. Create an array of 5 angles in degrees. Convert to radians and calculate sin and cos using np.pi.


```python
angles_deg = np.array([0, 30, 45, 60, 90])

# Convert degrees to radians using np.pi
angles_rad = angles_deg * (np.pi / 180)

# Calculate sine and cosine of each angle in radians
sin_values = np.sin(angles_rad)
cos_values = np.cos(angles_rad)

print("Angles in degrees:", angles_deg)
print("Angles in radians:", angles_rad)
print("Sine values:", sin_values)
print("Cosine values:", cos_values)
```

## 19) 19. Given an array with float values, convert it to integer using astype and calculate the sum before and after conversion.


```python
float_array = np.array([1.5, 2.7, 3.9, 4.2, 5.8])

# Calculate sum before conversion
sum_before = np.sum(float_array)

# Convert to integer using astype
int_array = float_array.astype(int)

# Calculate sum after conversion
sum_after = np.sum(int_array)

print("Original float array:", float_array)
print("Sum before conversion:", sum_before)
print("Converted integer array:", int_array)
print("Sum after conversion:", sum_after)
```

## 25. Use np.where to find positions of all elements divisible by 3 in a randomly generated 5x5 matrix.


```python
matrix = np.random.randint(1, 51, (5, 5))
print("Matrix:\n", matrix)

# Find positions where elements are divisible by 3
positions = np.where(matrix % 3 == 0)

print("Positions (row, column) of elements divisible by 3:")
for i in range(len(positions[0])):
    print(f"({positions[0][i]}, {positions[1][i]})")
```

## 26. Create a 2D array of shape (6,4). Find:

```python
arr_6x4 = np.random.randint(0, 100, (6,4))
mean_cols = np.mean(arr_6x4, axis=0)
std_all = np.std(arr_6x4)
median_rows = np.median(arr_6x4, axis=1)

print(mean_cols)
print(std_all)
print(median_rows)
```
