### Examine the distribution of unique categories in categorical columns

```python
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

## Univariate Analysis

- Univariate analysis means studying one column (variable) at a time to understand its data pattern.

### In this code —
It creates count plots (bar graphs) for each categorical column to see how often each value appears, helping to quickly understand the distribution of data in every column.

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

## Bivariate Analysis

- Bivariate analysis means studying the relationship between two variables at the same time.
- Bivariate analysis → means looking at 2 columns together to see how they are connected.

## In your code:

- x=predictor → first column (like GenderId, Occupation, etc.)

- hue='Nationality' → second column (Nationality)

- So this graph shows, for each category in the first column, how many people belong to each nationality.

- You use bivariate analysis to compare two columns and understand their relationship.

## In this code —

- It shows how each column (like GenderId, Occupation, etc.) changes with respect to another column — “Nationality.”
- x=predictor → the first variable (each column in loop)
- hue='Nationality' → the second variable used for comparison (adds color by nationality)
- So, we use bivariate analysis to see how one variable affects or relates to another.

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

# HIstplot of value counts for different Occupation
- A histogram shows how many times different values appear (the count). Here the code makes a histogram for each categorical column (except "Occupation"), so you can see the distribution (which categories are common or rare).

### What the code does (one-sentence summary):
- For every column in categorical_cols (but skipping "Occupation"), it draws a histogram showing how often each value appears, labels the axes and shows the plot.
- 

```python
for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

## Numerical Analysis
- Numerical analysis means studying columns that have numbers (like income, balance, loans, etc.) to understand how their values are spread — e.g., what range they have, which values are common, and the overall pattern.



```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

## Heatmaps

- A heatmap is a colorful chart that shows how strongly different numerical columns are related to each other.
- It helps us see which numbers increase or decrease together — this relationship is called correlation.

  

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Insights of EDA:
- The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
