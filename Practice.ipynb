## 1) Print the Prime numbers between 1,n. Take the input "n" from the user.

```python
n = int(input())
for num in range(2, n+1):
    is_prime= True
    for i in range(2, int(num**0.5)+1):
        if num % i== 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```

## 2) Given a list of integers and a target value, return indices of the two numbers that add up to the target.

```python
nums = [1, 3, 4, 5, 6, 7, 8]
target = 4

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == target:
            print([i, j])
```

## 3) Reverse a number n (using while loop)

```python
n = int(input())
reversed_num = 0
while n>0:
    digit = n % 10
    reversed_num = reversed_num * 10 + digit
    n = n//10
print(reversed_num)
```

```python
nums = eval(input("Enter list: "))  # Example: [0, 1, 0, 3, 12]

result = []         # To store non-zero elements
zero_count = 0      # Count how many zeros are there

for num in nums:
    if num != 0:
        result.append(num)   # Keep non-zero
    else:
        zero_count += 1      # Count zeros

# Add all zeros at the end
result.extend([0] * zero_count)

print(result)
```
