## Types of Charts and Plots
- Bar Plot → Shows data using bars (for comparing categories).

- Count Plot → Shows how many times each category appears (counts).

- Histogram → Shows how data is spread (distribution).

- Line Plot → Shows trends over time.

- Scatter Plot → Shows relationship between two variables.

- Pie Chart → Shows part-to-whole (percentage).

## Scaling
Changing data to a common scale (example: making numbers between 0 and 1).

Why? So all features are treated equally in graphs or models.

## Encoding

Turning categories into numbers.

#### Example:

Colors → Red = 1, Blue = 2, Green = 3.

#### Are Scaling & Encoding part of Data Visualization?

No, they are data preprocessing steps.

We do them before visualization or modeling.

# What is a Scatter Plot?

A scatter plot is a type of data visualization that shows the relationship between two numerical variables using points on a graph.

X-axis → represents one variable

Y-axis → represents another variable

Each dot (point) → represents one observation (data point)

## Why use it?

To see if two variables have a relationship (positive, negative, or no relation).

To detect patterns, clusters, and outliers in data.

#### Example

Suppose we have data of students:

#### Hours Studied vs Exam Score

Scatter plot will show:

If points go upward from left to right → Positive correlation (more study, higher score)

If points go downward → Negative correlation

If points are random → No correlation

### Here, points roughly form an upward trend → positive correlation.

```python
a = [40, 60, 30, 35, 15, 36]
b = [15, 5, 20, 21, 30, 19]
```
