## Question1 find the area of circle

```python
def area_of_circle(radius):           # Defines a function named 'area_of_circle' that takes 'radius' as input
    pi = 3.14                         # Sets the value of pi as 3.14
    area = pi * radius * radius      # Calculates the area using formula π * r * r
    return area                      # Returns the calculated area
r = float(input("Enter the radius: "))  # Takes input from the user, converts it to float, and stores in 'r'
result = area_of_circle(r)           # Calls the function with user's input and stores the result
print("Area of the circle is:", result)  # Prints the final result to the screen
```

## Question2 check even or odd using the fucntions

```python
def check_even_odd(number):            # Define a function that takes a number as input
    if number % 2 == 0:                # Check if the number is divisible by 2 (no remainder)
        print("The number is even")    # If true, print that it's even
    else:                              # If not divisible by 2
        print("The number is odd")     # Print that it's odd
num = int(input("Enter a number: "))   # Ask the user to enter a number and convert it to integer
check_even_odd(num)                    # Call the function and pass the entered number to check
```

## questions3 even and odd with list comprehension

```python
def sep_even_odd(lst):                     # Define a function that takes a list
    e = [n for n in lst if n % 2 == 0]     # Make a list 'e' with even numbers only
    o = [n for n in lst if n % 2 != 0]     # Make a list 'o' with odd numbers only
    return e, o                            # Return both even and odd lists
nums = [1, 2, 3, 4, 5, 6]                  # A sample list of numbers
ev, od = sep_even_odd(nums)               # Call the function and get even, odd numbers
print("Even:", ev)                        # Print the even numbers
print("Odd:", od)                         # Print the odd numbers
```

## ii) Filter Function:-
- The filter() function is used to select items from a list (or any iterable) based on a condition.

```python
l = [1, 2, 3, 4, 5, 67, 8]
evens= list(filter(lambda a: a%2==0, l))
print(evens)
```

```python
l = ["Apple", "orange", "Grapes", "WaterMelons"]
greater = list(filter(lambda a: len(a)>7, l))
print(greater)
```

## iii) reduce Function:-
- It takes two values at a time, does something (like add, multiply), then keeps using the result with the next value.
- It keeps going until only one final result is left.

```python
from functools import reduce
l = [1, 2, 3, 4, 5, 6]
red = reduce(lambda a, b: a + b, l)
print(red)
```

```python
from functools import reduce

nums = [1, 2, 3, 4]
result = reduce(lambda a, b: a + b, nums)
print(result)           # It does: 1+2=3, 3+3=6, 6+4=10
```

## Find the maximum value using reduce()

```python
from functools import reduce

nums = [5, 10, 3, 8]
maximum = reduce(lambda a, b: a if a > b else b, nums)
print(maximum)
```

# Scope Of Variables:-
- Variable scope means where in your code a variable can be used or accessed.

### 1) Global Variable:-
- A variable created outside all functions.
- Can be used anywhere in the program.

```python
name = "Sindhu"                  # Global variable 'name' declared outside all functions
def show():                      
    print(name)                  # Accesses and prints the global variable 'name'
show()                           # Calls the 'show' function
```

### 2) Local variable:-
- A variable created inside a function.
- Can only be used within that function.

```python
def greet():                      # Defines a function named 'greet'
    name = "Sindhu"              # Creates a local variable 'name' with value 'Sindhu'
    print(name)                  # Prints the value of 'name'
greet()                          # Calls the 'greet' function
```

```python
x = 5           ## Global variable
def ab():
    y= 10       ## Local variable
    print(x)
    print(y)
ab()
```

# Recursion:-
- Recursion means calling a fcuntion in items of itself.
- Recursion is when a function calls itself.
- It keeps calling itself until a stopping point (called base case).
##### How it works ?
- Function repeats by calling itself again and again with smaller input.
- It stops when a condition is met (base case), and then the answers go backward.

## Syntax:-

def function_name(parameters):
    if base_case_condition:
        return base_case_condition
    else:
        return function_name(smaller_problem)

#### n-1 helps the number go smaller each time, and finally reach the base case:

### eg1:-

```python
def fact(n):
    if n == 1:           # Base case: stop when n is 1
        return 1
    return n * fact(n-1) # Recursive call: n * factorial of (n-1)

print(fact(5))
```

### eg2:-

```python
def fact(n):
    if n == 0 or n==1:
        return 1
    return n * fact(n-1)
print(fact(5))
```

### eg3:-

#### Sum of Natural Numbers

```python
def sum(n):
    if n == 1:
        return 1
    return n+sum(n-1)
print(sum(5))
```

## Fibonacci Series Using Recursion:-

```python
def fibonacci(n):
    if n ==0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(10))
```

### Reverse a String Using Recursion

```python
def reverse_string(n):
    if n == "": 
        return n
    return reverse_string(n[1:]) + n[0]
print(reverse_string("Sindhu"))
```

```python
def sum(n):
    if n == 0:
        return 0
    return n+sum(n-1)
print(sum(5))
```
