```python
n = int(input())

for i in range(1, n + 1):
    row = ""
    for j in range(1, i + 1):
        row += str(j) + " "              # Make one line of numbers

    print(row.strip().center(n * 2))     # Remove extra space and center
```

```python
n = int(input("Enter an odd number: "))

for i in range(1, n + 1, 2):
    spaces = (n - i) // 2
    print(" " * spaces + "*" * i)

for i in range(n - 2, 0, -2):
    spaces = (n - i) // 2
    print(" " * spaces + "*" * i)
```

```python
text = input()              # Take input from user
text = text.lower()         # Convert to lowercase

count = 0                   # To count vowels

for char in text:           # Go through each character
    if char in "aeiou":     # Check if it's a vowel
        count += 1          # Increase count

print(count)                # Print the total number of vowels
```

#### Find maximum number in a list without using max()

```python
lst = [9, 2, 15, 4]
max_num = lst[0]
for i in lst:
    if i > max_num:
        max_num = i
print(max_num)
```
