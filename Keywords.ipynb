# Variables

- are used to store values
- place holders
## Syntax:-
- variable name = value
eg:- x = 5 

## Rules to define a variable name:-
- Rule 1:- we can use alphabets(A-Z)/ (a-z) (or) Combination of capital & small letters as variable name
- Rule 2:- Never start variable name with numbers and dont take only number of a variable
- Rule 3:- No special characters as a variable name except underscore(-)
- Rule 4:- Spaces between the Variables name is not allowed
- Rule 5:- Never use Keywords as variable names
- eg:- True=5 true=5 # wrong
- Rule 6:- Python is case-sensitive
