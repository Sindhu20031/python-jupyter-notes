# Loops

## Why Loops?
- we use loops because sometimes we need to write code again and again, so we take the help of loops so that we can write the code with fewer lines
- To avoid writing the same code many times.
- It saves time and makes code shorter.

 

## What Loops?
- loops in python program is used to repeat set of instructions until the condition is fulfilled.
- For loop
- While loop
- nested for loop
- nested while loop
- if we get stuck in loops we have control statements such as pass, break, and continue

# range function or range generator:-

### eg1:-

```python
l1 = [1, 2, 3, 4, 4, 9]        # A list with 6 numbers
for i in range(4):             # Loop will run 5 times (i = 0 to 4)
    print(l1[i])                   # Print the current value of i
```

### eg2:-

### eg3:- Print even numbers from 10 to 18

### What range(1, 11, 2) means:
1 → starting number

11 → goes till 10 only (end-1)

2 → jump in steps of 2 (so it skips every second number)

### eg4:-

```python
for i in range(1, 11, 2):     # Start from 1, go till 10, step by 2
    print(i)                  # Print the current number
```

### Find the factorial of a number using for loop.

```python
num = int(input("Enter a number: "))
fact = 1                             # Start with 1
for i in range(1, num + 1):          # Loop from 1 to num
    fact = fact * i                  # Multiply each number
print("Factorial is:", fact)
```

### sum of numbers from 1 to 10 using for loop and range():

```python
total = 0
for i in range(1, 11):     # Loop from 1 to 10
    total = total + i      # Keep adding each number

print("Sum is:", total)
```

```python
sum = 0                    # Start with 0, this will store the total
for i in range(1, 11):     # Loop from 1 to 10
    sum += i               # Same as: sum = sum + i (add each number)
print("Sum is:", sum)      # Print the total sum
```

# Index Based iterations:-
- Index-based iteration means using a number (index) to access each item in a list by its position.

```python
x = [1, 2, 3, 4, 5]             # A list with some numbers
a = []                         # Create an empty list to store results
for i in range(len(x)):        # Loop through indexes of x (0 to 4)
    a.append(x[i] * 2)         # Multiply each item by 2 and add to list 'a'
print(a)                       # Print the final new list
```

### Given a list of student marks, print "Pass" if mark is 40 or above, else print "Fail".

```python
marks = [35, 60, 45, 22, 80]   # List of marks
ls1 = []                      # Empty list to store results
for i in marks:               
    if i >= 40:
        ls1.append("Pass")    # Add "Pass" to ls1
    else:
        ls1.append("Fail")    # Add "Fail" to ls1

print(ls1)   
```

# nested for loop
- we can take many loops inside loop 

#### Syntax:-

#### for outer in outer_iterable:
       # for inner in inner_iterable:
        # Code block
        print(outer, inner)

## eg1:-

```python
a = ["red", "orange", "blue", "green"]       # List 'a' has some color names
b = ["apple", "apple", "grapes", "orange"]   # List 'b' has some fruit names
for i in a:                                  # Outer loop: goes through each color in list 'a'
    for j in b:                              # Inner loop: for each color, go through each fruit in list 'b'
        print(i, j)                          # Print the current color and fruit together
```

```python
for i in range(2):              # Outer loop: i will take values 0 and 1
    for j in range(10, 13):     # Inner loop: j will take values 10, 11, 12
        print(i, j)             # Print the current values of i and j
```

```python
letters = 'abc'
for i in letters:
    for j in letters:
        print(i+j)
```

```python
letters = ['a', 'b','c']
for i in letters:
    for j in letters:
        print(i+j)
```

```python
for i in range(1, 11):
    for j in range(10, 120)# Loop from 1 to 10
        print("1 x", i, "=", 1 * i)      # Print 1 multiplied by i
```

```python
for i in range(1, 11):                    
    for j in range(1, 11):         
        print(f"{i} x {j} = {i * j}")
    print()     
```

```python
for i in range(5):
    for j in range(i):
        print(" * ", end ='')
    print()
```

```python
for i in range(1, 6):            # Rows from 1 to 5
    for j in range(1, i + 1):    # Numbers from 1 to i
        print(j, end=' ')        # Print current number
    print() 
```

```python
for i in range(1, 6):                     # Rows from 1 to 5
    print('  ' * (5 - i), end='')         # Print spaces for alignment (2 spaces per gap)
    for j in range(1, i + 1):             # Print numbers from 1 to i
        print(j, end=' ')
    print()      
```

# Nested while loop

#### Syntax:-

##### while condition1:
    # outer loop code
- --------> while condition2:
- ----------->  # inner loop code

## eg:-

```python
a = 1
b = 5
while a <=5:
    while b<=10:
        print(a,b)
        b=b+1
        a=a+1
```

```python
i = 1                      # Start row from 1
while i <= 5:              # Loop until row 5
    j = 1                  # Start printing numbers from 1 in each row
    while j <= i:          # Print numbers from 1 to i
        print(j, end=' ')  # Print the number and stay on same line
        j += 1             # Move to next number
    print()                # Move to next line after finishing row
    i += 1  
```

## control statements for while loop
- Pass
- break
- continue

### Pass

```python
i = 1
while i< 5:
    print(i)
    pass
    i=i+1 
```

### break

```python
i = 1
while(i<= 5):
    print(i)
    if i ==3:
        break
    i = i+1
```

```python
i = 1
while i <= 5:
    if i == 3:
        break
    print(i)
    i = i + 1
```

### Continue:

```python
i = 1                       # Start with i = 1
while(i <= 5):              # Loop as long as i is 5 or less
    i = i + 1               # First, increase i by 1
                            # So now i becomes 2, then 3, then 4, etc.
    if i == 3:              # If i becomes 3,
        continue            # Skip the rest and go to next loop round
    print(i)                # Print the current value of i (but not if i was 3)
```
