##  What is Exception Handling?
- In Python, exception handling is used to catch errors so the program doesn't crash.
- We use try, except, else, and finally.

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except:
    print("Something went wrong!")
```

```python
try:
    a = int(input("Enter a number: "))
    b = 10 / a
    print(b)
except ZeroDivisionError:
    print("You cannot divide by zero!")
except ValueError:
    print("You must enter a number.")
```

```python
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid input!")
else:
    print("You entered:", num)
```

```python
try:
    f = open("test.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found.")
finally:
    print("Finished trying to read file.")
```
