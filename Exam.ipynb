```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Sindhu\Downloads\tips.csv")
df
```

## 1. Calculate the average total bill and average tip for Sundays

```python
avg_total_bill = df[df['day'] == 'Sun']['total_bill'].mean()
avg_tip = df[df['day'] == 'Sun']['tip'].mean()
print("Average total bill on Sunday:", avg_total_bill)
print("Average tip on Sunday:", avg_tip)
```

## 2. Find the highest tip amount given by a female customer during dinner on a Sunday.

```python
filtered = df[(df['sex'] == 'Female') & (df['time'] == 'Dinner') & (df['day'] == 'Sun')]
highest_tip = filtered['tip'].max()

print("Highest tip given by a female customer during dinner on Sunday:", highest_tip)
```

## 3. Create a new column named `tip_pct` that stores the tip percentage of the total bill.


```python
df['tip_pct'] = (df['tip'] / df['total_bill']) * 100
df
```

## 4. Create a new column named `bill_per_person` by dividing *total_bill* by the number of people (*size*).

```python
df['bill_per_person']=(df['size']/df['total_bill']) * 100
df
```

## 5. Add a new column called `date` containing dates starting from today’s date (assign one date per row).

```python
from datetime import date, timedelta
today = date.today()
dates = [today + timedelta(days = i) for i in range(len(df))]
df["date"] = dates
print(df.head(4))    
```

## 6. From the date column, extract the month number and store it in a new column called month.

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
print(df.head())
```

## 7. Standardize the values in the `sex` column by converting them to lowercase and then capitalizing the first letter (e.g., "male" → "Male")


```python
df['sex'] = df['sex'].str.lower().str.capitalize()
print(df['sex'].unique())
```

## 8. Create a new column named customer_type based on the number of people (size) at the table:

- Assign **"Family"** if size is 5 or more

- Assign **"Friends"** if size is 3 or 4

- Assign **"Lovers"** if size is exactly 2

- Assign **"Single"** if size is 1

```python
import numpy as np
conditions = [
    (df['size'] >=5),
    (df['size'].isin([3,4])),
    (df['size'] == 2),
    (df['size'] ==1)
]
values = ['Family', 'Friends', 'Lovers', 'Single']
df['customer_type'] = np.select(conditions, values)
print(df[['size', 'customer_type']].head(10))
```

## 9. Filter and display only those rows where the percentage of tip (tip / total_bill × 100) is greater than 20%.

```python
high_tips = df[df['tip_pct']>20]
print(high_tips.head())
```

## 10. Generate a `pivot table` that shows the average `total_bill` grouped by day (rows) and sex (columns).

```python
import pandas as pd
pivot = pd.pivot_table(df, values ='total_bill', index = 'day', columns='sex', aggfunc='mean')
print(pivot)
```

## 11. Identify and display the `outliers` in the `tip` column using the Interquartile Range (IQR) method.

```python
# Step 1: Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['tip'].quantile(0.25)
Q3 = df['tip'].quantile(0.75)

# Step 2: Calculate IQR
IQR = Q3 - Q1

# Step 3: Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Filter out outliers
outliers = df[(df['tip'] < lower_bound) | (df['tip'] > upper_bound)]

print("Outliers in tip column:")
print(outliers)
```

## 1. Generate a NumPy array of 100 random integers between 10 and 100.
- Find the mean, median, variance, and standard deviation.

- Count how many numbers are above the mean.

```python
import numpy as np
arr = np.random.randint(10, 101, size = 100)
print(arr)
```

```python
mean_val = np.mean(arr)
median_val = np.median(arr)
variance_val = np.var(arr)
std_dev_val = np.std(arr)
count = np.sum(arr > mean_val)
print("Mean:", mean_val)
print("Median:", median_val)
print("Variance:", variance_val)
print("Standard Deviation:", std_dev_val)
print("Count above mean:", count)
```
