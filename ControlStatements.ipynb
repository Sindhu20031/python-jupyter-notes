# 3) if-elif-else
- if-elif-else lets your program pick one action out of many—first it checks the if, then each elif, and if none match, it does the else.

## Syntax:-

#### if condition1:
    # code when condition1 is true
#### elif condition2:
    # code when condition2 is true
#### else:
    # code when none of the above are true

```python
score = int(input("Enter your score: "))  
if score >= 90:
    print("A grade")
elif score >= 75:
    print("B grade")
else:
    print("C grade or below")
```

```python
x = int(input("Enter a number: "))  # get a number from the user
if x > 0:
    print("Positive")
elif x < 0:
    print("Negative")
else:
    print("Zero")
```

##  Find the largest of three numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:                 # a >= b` ensures a is not less than b, a >= c` ensures a is not less than c
    print("The largest number is", a)
elif b >= a and b >= c:
    print("The largest number is", b)
else:
    print("The largest number is", c)
```

## Check day of the Week

```python
day_num = int(input("Enter day number (1–7): "))  # 1=Monday … 7=Sunday

if day_num == 1:
    print("Monday")
elif day_num == 2:
    print("Tuesday")
elif day_num == 3:
    print("Wednesday")
elif day_num == 4:
    print("Thursday")
elif day_num == 5:
    print("Friday")
elif day_num == 6:
    print("Saturday")
elif day_num == 7:
    print("Sunday")
else:
    print("Invalid day number!")
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
d = int(input("Enter day number (1–7): "))

# days[d-1:d] is an empty list if d-1 is out of range, otherwise a one‑item list
if days[d-1:d]:
    print(days[d-1])    ## print the single day from the list
else:
    print("Invalid day number!")
```

## Simple Calculator using only if‑elif‑else

```python
a = int(input("Enter the Value:"))
b = int(input("Enter the Value:"))
op = input("Enter the Operators:")
if op == '+':
    print(a+b)
elif op == '-':
    print(a-b)
elif op == '*':
    print(a*b)
elif op =='/':
    print(a/b)
else:
    print("Invalid Operator")           
```

```python
num1 = float(input("Enter first number: "))
op   = input("Enter operation (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if op == "+":
    print("Result:", num1 + num2)
elif op == "-":
    print("Result:", num1 - num2)
elif op == "*":
    print("Result:", num1 * num2)
elif op == "/" and num2 != 0:
    print("Result:", num1 / num2)
elif op == "/" and num2 == 0:
    print("Error: Cannot divide by zero")
else:
    print("Invalid operation")
```

```python
num1 = float(input("Enter first number: "))
op   = input("Enter operation (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if op == "+":
    result = num1 + num2
elif op == "-":
    result = num1 - num2
elif op == "*":
    result = num1 * num2
elif op == "/" and num2 != 0:
    result = num1 / num2
elif op == "/" and num2 == 0:
    print("Error: Cannot divide by zero")
    result = None
else:
    print("Invalid operation")
    result = None
if result is not None:
    print("Result:", result)
```

## Simple Discount Calculator


```python
price = int(input("Enter the Price: "))
if price > 2000:
    discount = 20
elif price > 1000:
    discount = 10
else:
    discount = 5
print(f" discount given:{discount}%")
```

```python
price = int(input("Enter the price:"))

if price > 2000:
    print("price after discount:", price - price // 20)
elif price > 1000:
    print("price after discount:", price - price // 10)
else:
    print("price after discount:", price - price // 5)
```

```python
price = int(input("Enter the Price: "))

if price > 2000:
    discount = 20
elif price > 1000:
    discount = 10
else:
    discount = 5

discount_amount = price * discount / 100
final_price = price - discount_amount

print(f"Discount: {discount}%")
print(f"Discount amount: Rs{discount_amount}")
print(f"Final price after discount: Rs{final_price}")
```

# 4) nested if:-
- Nested if means writing an if statement inside another if, to check more conditions step by step.


### Syntax:-

```python
x = int(input("Enter the Value: "))         # Take input from user and convert to int
if x > 0:                                   # Check if number is positive
    print("Yes")                            # If yes, print Yes
    if x <= 50:                             # Check if it's 50 or below
        print("Value is less than 50")      # If true, print this message
        if x % 2 == 0:                      # Check if number is even
            print("Is a Even Number")       # If even, print this
```

```python
x = int(input("Enter the value: "))          # Take number input from user
if x > 0:                                    # Check if number is positive
    print("yes")                             # If yes, print "yes"
    if x <= 50:                              # Check if number is less than or equal to 50
        print("value is less than 50")       # Print this if true

        if x % 2 == 0:                       # Check if number is even
            print("Is a even number")        # If even, print this
        else:
            print("is a odd number")         # If not even, print odd
    else:
        print("value is greater than 50")    # If not ≤ 50, then it is > 50

# If x is not greater than 0, nothing will be printed
```
