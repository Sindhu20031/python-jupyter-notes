## Mini Task 1: Rock-Paper-Scissors

````python
import random
choices = ["rock", "paper", "scissors"]
emojis = ["\U0001FAA8", "\U0001F4C4", "\u2702\ufe0f"]

user_input = input("Choose rock, paper, or scissors: ").lower()

if user_input not in choices:
    print("Invalid choice. Please choose rock, paper, or scissors.")
else:
    comp_index = random.randint(0, 2)
    comp_choice = choices[comp_index]
    
    user_index = choices.index(user_input)
    
    print("You chose:", user_input, emojis[user_index])
    print("Computer chose:", comp_choice, emojis[comp_index])
    
    if user_index == comp_index:
        print("It's a draw!")
    elif (user_index == 0 and comp_index == 2) or \
         (user_index == 1 and comp_index == 0) or \
         (user_index == 2 and comp_index == 1):
        print("Yay! You win")
    else:
        print("You lose!")
````

## Mini Task 2: ATM Cash Withdrawal

###### #### Goal:

Simulate a simple ATM cash withdrawal system using nested `if` statements.

#### Try to use these concepts:

1. `if`, `else`, and **nested `if`**
2. User `input()`
3. Variables for **balance** and **withdrawal amount**
4. Basic arithmetic and condition checks

#### What to Do?

1. Set a fixed account balance, e.g., `balance = 5000`

2. Ask the user to enter the amount to withdraw

3. Check the following using nested `if`:

    a) Is the withdrawal amount less than or equal to the balance?
   
    b) Is the amount a multiple of 100?

4. If both conditions are satisfied, deduct the amount and show the remaining balance

5. Else, show an appropriate error message

#### Hints:

* Use the **modulus operator** `%` to check if the amount is a multiple of 100
* Nest the second `if` inside the first one
* Always handle the "else" case for both checks with user-friendly error messages

#### Example Output:

```
Enter amount to withdraw: 1200  
Withdrawal successful! Remaining balance: 3800
```

```
Enter amount to withdraw: 1250  
Error: Amount must be a multiple of 100.
```

```
Enter amount to withdraw: 6000  
Error: Insufficient balance.
```



````python
balance = 5000                             # Fixed account balance

amount = int(input("Enter amount to withdraw: "))   # Take input from user

if amount <= balance:                     # Check if balance is enough
    if amount % 100 == 0:                 # Check if amount is multiple of 100
        balance = balance - amount        # Subtract amount from balance
        print("Withdrawal successful!")
        print("Remaining balance:", balance)
    else:
        print("Error: Amount must be a multiple of 100.")
else:
    print("Error: Insufficient balance.")
````

## Mini Task-3: Number Guessing Game

##### Goal:
Build a number guessing game where the user tries to guess a randomly generated number between 1 and 100. After each guess, give hints like "Too high", "Too low", or "Correct!".

#### Try to use these concepts:
1) `if`, `elif`, `else`  
2) `input()` and `int()` conversion  
3) `random.randint()`  
4) Loops (optional for bonus)  
5) Variables and comparisons  

#### What to Do?
1) Generate a random number between 1 and 100 using `random.randint(1, 100)`  
2) Ask the user to guess the number  
3) After each guess, give feedback:  
 a) "Too high" if the guess is more than the number  
 b) "Too low" if the guess is less than the number  
 c) "Correct!" if the guess is equal  

####  Bonus Challenge:
- Give the user **only 5 attempts**  
- After 5 wrong guesses, end the game and reveal the correct number  

#### Hints:
- Convert user input to integer using `int()`  
- Keep track of attempts using a counter  
- Use a `while` loop or `for` loop for attempts (if doing the bonus part)  

#### Example Output:
```
example 1:
Guess the number (between 1 and 100): 50
Too low. Try again.

example 2:
Guess the number (between 1 and 100): 75
Too high. Try again.

example 3:
Guess the number (between 1 and 100): 63
Correct!!You guessed it right.
```

#### Example Output (for bonus section):
```
Guess the number (between 1 and 100): 50
Too low. Try again.

Guess the number (between 1 and 100): 75
Too high. Try again.

Guess the number (between 1 and 100): 63
Correct!!You guessed it in 3 attempts.
```



````python
import random

secret = random.randint(1, 100)     # Random number between 1 and 100
attempts = 0                        # To count number of tries

while attempts < 5:                 # Only allow 5 guesses
    guess = int(input("Guess the number (1 to 100): "))
    attempts += 1                  # Count this attempt

    if guess == secret:
        print("Correct!! You guessed it in", attempts, "attempts.")
        break
    elif guess < secret:
        print("Too low. Try again.")
    else:
        print("Too high. Try again.")

# If not guessed in 5 tries
if guess != secret:
    print("Out of attempts! The number was", secret)
````

## Mini Task 4: Speed Limit Fine Calculator

###### Goal:
Build a Speed Fine Calculator that checks how fast the user is driving and calculates the fine based on how much they exceed the speed limit.
---

#### Try to use these concepts:
1) if, elif, else

2) User input and type conversion

3) Arithmetic operations (subtraction)

4) Comparisons and conditions

#### What to Do?
1) Set the speed limit (e.g., 60 km/h)

2) Ask the user to enter their current speed

3) Calculate how much they exceeded the limit

4) Use conditions to print:
   - No fine (if within or at limit)
   - ₹500 fine (if 1–10 km/h over)
   - ₹1000 fine (if 11–30 km/h over)
   - ₹2000 fine + License Warning (if more than 30 km/h over)

#### Speed Limit Setup
`speed_limit = 60`  
`user_speed = int(input("Enter your speed: "))`  
`excess = user_speed - speed_limit`  

#### Hints
1) If excess <= 0 → No fine

2) If 1 <= excess <= 10 → ₹500 fine

3) If 11 <= excess <= 30 → ₹1000 fine

4) If excess > 30 → ₹2000 fine and show a warning

Example Output  
```
example 1:
      You were driving at: 65 km/h  
      You are 5 km/h over the limit. Fine: ₹500  
example 2:
      You were driving at: 85 km/h  
      You are 25 km/h over the limit. Fine: ₹1000  
example 3:
      You were driving at: 95 km/h  
      You are 35 km/h over the limit. Fine: ₹2000  
       Warning: Your license is at risk!  
example 4:
      You were driving at: 58 km/h  
      You're within the speed limit. No fine!
```

````python
speed_limit = 60                              # Fixed speed limit
user_speed = int(input("Enter your speed: ")) # Ask user for current speed

excess = user_speed - speed_limit             # Calculate how much over the limit

if excess <= 0:
    print("You're within the speed limit. No fine!")
elif excess <= 10:
    print("You are", excess, "km/h over the limit. Fine: ₹500")
elif excess <= 30:
    print("You are", excess, "km/h over the limit. Fine: ₹1000")
else:
    print("You are", excess, "km/h over the limit. Fine: ₹2000")
    print("Warning: Your license is at risk!")
````

## Mini Task 5: Loan Approval System

#### Goal:
Build a loan approval system that decides whether a loan application is approved, conditionally approved, or rejected based on credit score, income, and debt status

#### Try to Use These Concepts:
1. `if`, `elif`, `else`  
2. Logical operators: `and`, `or`, `not`  
3. User input and type conversion  
4. Nested conditions or combined checks

#### What to Do?:
1. Ask the user if they are currently in debt (`yes` or `no`)  
2. If the user **is in debt**, reject the application immediately  
3. Otherwise, ask for:
   - Credit score (integer)
   - Monthly income (integer in ₹)

4. Apply the following approval logic:
   - If **credit score ≥ 750** and **income ≥ ₹50,000** → **Loan Approved**   
   - Else if **credit score ≥ 650** and **income ≥ ₹35,000** → **Approved with Guarantee**
   - Else → **Loan Rejected**


#### Hints:
- Convert credit score and income to integers using `int()`  
- Convert debt status input to lowercase using `.lower()` to handle variations  
- Always check for debt status **before** checking other conditions

#### Example Input and Output


Case 1: Loan Approved
```

Are you currently in debt? no
Enter your credit score: 765
Enter your monthly income: 68000
Loan Approved
Congratulations! You meet all criteria.

```

Case 2: Approved with Guarantee
```

Are you currently in debt? no
Enter your credit score: 695
Enter your monthly income: 42000
Approved with Guarantee
Note: You need a guarantor or collateral for this loan.

```

Case 3: Loan Rejected (Low Score)
```

Are you currently in debt? no
Enter your credit score: 610
Enter your monthly income: 55000
Loan Rejected
Reason: Credit score too low.

```

Case 4: Loan Rejected (Low Income)
```

Are you currently in debt? no
Enter your credit score: 710
Enter your monthly income: 32000
Loan Rejected
Reason: Monthly income below required threshold.

```

Case 5: Automatically Rejected Due to Existing Debt
```

Are you currently in debt? yes
Loan Rejected
Reason: Applicant currently in debt.

```


````python
debt = input("Are you currently in debt? (yes/no): ").lower()

if debt == "yes":
    print("Loan Rejected")
    print("Reason: Applicant currently in debt.")
else:
    credit = int(input("Enter your credit score: "))
    income = int(input("Enter your monthly income (₹): "))

    if credit >= 750 and income >= 50000:
        print("Loan Approved")
        print("Congratulations! You meet all criteria.")
    elif credit >= 650 and income >= 35000:
        print("Approved with Guarantee")
        print("Note: You need a guarantor or collateral for this loan.")
    elif credit < 650:
        print("Loan Rejected")
        print("Reason: Credit score too low.")
    elif income < 35000:
        print("Loan Rejected")
        print("Reason: Monthly income below required threshold.")
````

## Mini Task 6: Traffic Light Behavior Simulation

#### Goal:
Create a Python program that tells a driver what to do based on the traffic light color and their intended direction.

#### Try to Use These Concepts:
1. `if`, `elif`, `else`  
2. **Nested `if` statements**  
3. User input and `.lower()` for clean comparison

#### What to Do:
1. Ask the user to enter the **traffic light color** (`red`, `yellow`, or `green`)  
2. Ask the user to enter their **intended direction** (`left`, `right`, or `straight`)  
3. Use nested `if-else` statements to decide and print the vehicle's behavior


#### Suggested Rules:
- **Red Light**  
  - All directions must stop

- **Yellow Light**  
  - All directions: slow down or prepare to stop

- **Green Light**  
  - If going **straight** or turning **right** → Go  
  - If turning **left** → Ask: “Is there oncoming traffic? (yes/no)”  
    - If yes → Wait  
    - If no → Turn left


#### Hints:
- Use `.lower()` on inputs to handle case differences  
- Use nested `if` inside the green light logic for left turn decision  
- No need to validate all inputs — focus on logic practice


#### Example Input and Output:

Case 1: Green Light, Going Straight
```

Enter traffic light color: green
Enter direction (left/right/straight): straight
Go ahead.

```

Case 2: Yellow Light, Turning Right
```

Enter traffic light color: yellow
Enter direction (left/right/straight): right
Slow down and prepare to stop.

```

Case 3: Green Light, Turning Left with Oncoming Traffic
```

Enter traffic light color: green
Enter direction (left/right/straight): left
Is there oncoming traffic? yes
Wait for oncoming traffic to clear.
```

Case 4: Green Light, Turning Left with No Oncoming Traffic
```

Enter traffic light color: green
Enter direction (left/right/straight): left
Is there oncoming traffic? no
You may turn left.

```

Case 5: Red Light, Going Straight
```

Enter traffic light color: red
Enter direction (left/right/straight): straight
Stop.

```


````python
light = input("Enter traffic light color: ").lower()
direction = input("Enter direction (left/right/straight): ").lower()

if light == "red":
    print("Stop.")

elif light == "yellow":
    print("Slow down and prepare to stop.")

elif light == "green":
    if direction == "left":
        traffic = input("Is there oncoming traffic? (yes/no): ").lower()
        if traffic == "yes":
            print("Wait for oncoming traffic to clear.")
        else:
            print("You may turn left.")
    else:
        print("Go ahead.")
````
