## twin prime numbers

```python
for num in range(2, 15):
    # check if num is prime
    prime1 = True
    for i in range(2, num):
        if num % i == 0:
            prime1 = False
            break

    # check if num + 2 is also prime
    prime2 = True
    for j in range(2, num + 2):
        if (num + 2) % j == 0:
            prime2 = False
            break

    # if both are prime, print as twin prime
    if prime1 and prime2:
        print(num, "and", num + 2, "are twin primes")
```

## Armstrong Number Check

```python
num = int(input("Enter a number: "))
sum = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp = temp // 10

if sum == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

### Given a list of integers, write a program to find the number closest to 0.
### If there are multiple numbers equally close to 0, return the larger number.
### The input will be a list like [-4, -2, 1, 4, 8].
### The output will be a single integer that is closest to 0 based on the rules.

```python
nums = [-4, -2, 1, 4, 8]
closest = nums[0]
for n in nums:
    if abs(n) < abs(closest):
        closest = n
    elif abs(n) == abs(closest) and n > closest:
        closest = n
print(closest)
```

### Each digit from the number is alternately added and subtracted, starting from positive for the first digit.

```python
n = input()      # Get number as a string
result = 0       # Start with 0
i = 0
while i < len(n):
    digit = int(n[i])   # Take one digit at a time
    if i % 2 == 0:      # If index is even, add
        result += digit
    else:               # If index is odd, subtract
        result -= digit
    i += 1              # Move to next digit
print(result)           # Print the final answer
```

## 2. Write a function that returns the nth Fibonacci number using memoization.

```python
def fibo(n, memo={}):
    if n in memo:  # if already calculated, return it
        return memo[n]
    if n <= 1:     # base cases
        return n
    memo[n] = fibo(n-1, memo) + fibo(n-2, memo)  # store in memo
    return memo[n]

print(fibo(5))  
print(fibo(10))
```

```python
def fibo(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a
fibo(10)
```

## 3) Create a function to return the longest substring without repeating characters.

```python
def longest_substring(s):
    start = 0
    max_len = 0
    used_chars = {}
    
    for i, char in enumerate(s):
        if char in used_chars and start <= used_chars[char]:
            start = used_chars[char] + 1
        used_chars[char] = i
        max_len = max(max_len, i - start + 1)
    
    return max_len

print(longest_substring("abcabcbb"))  
print(longest_substring("bbbbb"))     
print(longest_substring("pwwkew"))    
```

## 6) Implement a function that rotates a 2D matrix 90 degrees clockwise.

```python
def rotate(matrix):
    n = len(matrix)
    res = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            res[j][n-1-i] = matrix[i][j]
    for row in res:
        print(row)

rotate([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])
```
